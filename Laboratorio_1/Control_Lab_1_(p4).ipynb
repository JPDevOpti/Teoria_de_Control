### --- Sección 2: Álgebra Lineal con NumPy ---

```python
A = np.array([[3, 2], [1, 4]])
B = np.array([5, 6])

# Resolver Ax = B
x = np.linalg.solve(A, B)
print("Solución de Ax = B:", x)
```

```python
# Ejercicio: Calcular los valores propios y vectores propios de la matriz A

# Definir la matriz A
A = np.array([[4, -2, 1], 
              [1, 1, -1], 
              [3, -1, 2]])

# Calcular valores propios y vectores propios
valores_propios, vectores_propios = np.linalg.eig(A)

# Mostrar los resultados
print("Valores propios:")
print(valores_propios)

print("\nVectores propios:")
print(vectores_propios)
```

```python
# Ejercicio: Resolver un sistema de ecuaciones lineales usando scipy.linalg

# Definir la matriz de coeficientes A y el vector de términos independientes b
A = np.array([[3, 2, -1], 
              [2, -2, 4], 
              [-1, 0.5, -1]])

b = np.array([1, -2, 0])

# Resolver el sistema de ecuaciones Ax = b
x = scipy.linalg.solve(A, b)  # Llamamos a solve usando scipy.linalg.solve

# Mostrar la solución
print("Solución del sistema:")
print(x)
```

```python
# Ejercicio: Multiplicar los polinomios P y Q

# Definir los coeficientes de los polinomios P(x) y Q(x)
# P(x) = 3x² + 2x + 1
# Q(x) = x² - 4x + 2
P = np.array([3, 2, 1])
Q = np.array([1, -4, 2])

# Multiplicar los polinomios
Producto = np.polymul(P, Q)

# Mostrar el resultado
print("Polinomio resultante (coeficientes):")
print(Producto)
```

```python
# Ejercicio: Evaluar P(x) en x = 2

# Definir los coeficientes del polinomio P(x)
# P(x) = 3x² + 2x + 1
P = np.array([3, 2, 1])

# Valor de x donde se evaluará el polinomio
x = 2

# Evaluar el polinomio en x = 2
resultado = np.polyval(P, x)

# Mostrar el resultado
print(f"P({x}) =", resultado)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los coeficientes de los polinomios P(x) y Q(x)
P = np.array([3, 2, 1])  # P(x) = 3x² + 2x + 1
Q = np.array([1, -4, 2])  # Q(x) = x² - 4x + 2

# Crear un rango de valores de x en el intervalo [-5, 5]
x_vals = np.linspace(-5, 5, 100)

# Evaluar los polinomios en los valores de x
y_P = np.polyval(P, x_vals)
y_Q = np.polyval(Q, x_vals)

# Graficar los polinomios
plt.plot(x_vals, y_P, label="P(x)")
plt.plot(x_vals, y_Q, label="Q(x)", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Gráfica de P(x) y Q(x)")
plt.grid(True)
plt.show()
```
